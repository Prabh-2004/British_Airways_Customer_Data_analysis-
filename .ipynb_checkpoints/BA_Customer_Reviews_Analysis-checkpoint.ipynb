```python
sentences = df['reviews']
sentences = list(sentences)
```

```python
import re
import nltk
# nltk.download('english')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer

reviews_preprocessed = []

for i in range(len(sentences)):
    sentences[i] = re.sub('[^a-zA-Z]', ' ', sentences[i])
    sentences[i] = sentences[i].lower()
    words = nltk.word_tokenize(sentences[i])
    # stemmer = PorterStemmer()
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(word, 'v') for word in words if word not in set(stopwords.words('english'))]
    sentences[i] = ' '.join(words)
    reviews_preprocessed.append(sentences[i])
```

```python
reviews_preprocessed
```

```python
len(reviews_preprocessed)
```

## Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X = cv.fit_transform(reviews_preprocessed)
```

```python
X = X.toarray()
```

```python
X
```

```python
X.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(df['recommend'])
```

```python
y
```

## Modelling

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

y_train_pred = classifier.predict(X_train)

print(accuracy_score(y_train, y_train_pred))
print(confusion_matrix(y_train, y_train_pred))
```

```python
y_test_pred = classifier.predict(X_test)

print(accuracy_score(y_test, y_test_pred))
print(confusion_matrix(y_test, y_test_pred))
```

Out of Vocabulary words in the test set caused a sharp drop in test accuracy.
