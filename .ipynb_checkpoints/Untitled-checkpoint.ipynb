```python
with open("file.html", "r") as f:
    html = f.read()
```

```python
# tables = pd.read_html(response.text)

soup = BeautifulSoup(html, 'html.parser')
print(soup.prettify())
```

```python
review = []
heading = soup.find_all("h2", class_="text_header")

for h2 in heading:
    review.append(h2.get_text())

review
```

```python
recommend = []
recommendation = soup.find_all('td')

for i in recommendation:
    if (i.get('class') == ['review-value', 'rating-yes']) | (i.get('class') == ['review-value', 'rating-no']):
        recommend.append(i.get_text())
```
