```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
review = []
recommend = []
country = []
date = []
seat_type = []
```

```python
response = requests.get("https://www.airlinequality.com/airline-reviews/british-airways/page/35/?sortby=post_date%3ADesc&pagesize=100")
html = response.text
soup = BeautifulSoup(html, "html.parser")

for item in soup.find_all("h2", class_="text_header"):
    review.append(item.get_text())

for item in soup.find_all("td"):
    if (item.get('class') == ['review-value', 'rating-yes']) | (item.get('class') == ['review-value', 'rating-no']):
        recommend.append(item.get_text())

for item in soup.find_all("h3"):
    country.append(item.span.next_sibling.text.strip(" ()"))

for item in soup.find_all("time"):
    date.append(item.get('datetime'))

for item in soup.find_all("td"):
    if item.get('class') == ['review-rating-header', 'cabin_flown']:
        seat_type.append(item.next_sibling.get_text())
```

```python
print(len(review))
print(len(recommend))
print(len(country))
print(len(date))
print(len(seat_type))
```
